```python
import numpy as np

def hambac2(a,b,c,x):
  f=a*x**2+b*x+c
  return f

def hambac3(a,b,c,d,x):
  f=a*x**3+b*x**2+c*x+d
  return f

def hambac1(a,b,x):
  f=a*x+b
  return f

x=np.linspace(-5,15,1000) 

y2=hambac2(-2,1,5,x)
y1=hambac1(1,3,x)
y3=hambac3(1,3,5,2,x)


import matplotlib.pyplot as plt

%matplotlib inline 
fig,ax=plt.subplots()

ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
ax.set_xlabel("Trục hoành - x")
ax.set_ylabel("Trục tung - y")
ax.plot(x, y1,label=r'$y=x+3$')
ax.plot(x, y2,label=r'$y=-2x^2 + 1x+5$')
ax.plot(x, y3,label=r'$y=x^3+3x^2+5x+2$')
ax.legend()
plt.show()


```

```python
import numpy as np
from math import cos,sin,pi

def hamcos(x):
  f=np.cos(x/pi)
  return f

def hamsin(x):
  f=np.sin(x/3)
  return f

x=np.linspace(-10,10,1000) 
y1=hamcos(x)
y2=hamsin(x)

import matplotlib.pyplot as plt
%matplotlib inline 
fig,ax=plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlabel("Trục hoành - x")
ax.set_ylabel("Trục tung - y")
ax.plot(x, y1,label=r'$y=cos(x/pi)$')
ax.plot(x, y2,label=r'$y=sin(pi*x/3)$')
ax.legend()
plt.show()


```

```python
%matplotlib inline
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
x = np.linspace(start=-15,stop=15,num=1000)

y1 = hambac1(1,3,x)
y2 = hambac2(-2,1,5,x)
y3 = hambac3(1,3,5,2,x) 

y4 = hamcos(x)
y5 = hamsin(x)

ax1.plot(x, y1)
ax1.plot(x, y2)
ax1.plot(x, y3)

ax2.plot(x, y4)
ax2.plot(x, y5)

ax1.set_xlabel("Trục hoành - x")
ax1.set_ylabel("Trục tung - y")

ax2.set_xlabel("Trục hoành - x")
ax2.set_ylabel("Trục tung - y") 

ax1.plot(x, y1, label = r'$y=x+3$')
ax1.plot(x, y2, label = r'$y=-2x^{2} - 1x + 5$')
ax1.plot(x, y3, label = r'$y=x^{3} + 3x^{2} + 5x 2')

ax2.plot(x, y4, label = r'$y=cos(x/pi)$')
ax2.plot(x, y5, label = r'$y=sin((pi/3) * x)$')

ax1.set_title("Đồ thị phương trình bậc 1,2,3")
ax1.set_title("Đồ thị hàm cos, sin")

ax1.legend()
ax2.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def ham(x,y):
  f=x**2+y**2-np.cos(12*x)-np.cos(18*y)
  return f
x=np.linspace(-1,1,1000)
y=np.linspace(-1,1,1000)

x, y = np.meshgrid(x, y)
z=ham(x,y)
fig, ax =plt.subplots(subplot_kw={"projection":"3d"})

rosen_surf = ax.plot_surface(x, y, z,cmap=cm.gist_heat_r, linewidth=0, antialiased=False)

ax.set_zlim(-2,4)

fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
plt.show()








```
