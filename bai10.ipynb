```python
import numpy as np
from numpy import random
from math import sin,cos,log,sqrt
x=random.randint(1,10,size=(1,5))

def bai1(x):
    print(np.round(np.sin(x),2))
    print(np.round(np.cos(x),2))
    print(np.round(np.sqrt(x),2))
    print(np.round(np.log(x),2))
    y=random.randint(1,10,size=(3,5))
    print(np.round(np.cos(y),3))
    print(np.round(np.sin(y),3))
    print(np.round(np.tan(y),3))
    print(np.round(np.sqrt(y),3))    
    print(np.round(np.log(y),3))
    
def bai2(x):
    print(np.linspace(-100,100,1000))
    f=3*x**2-2*x-4
    print(cos(f))
    print(sin(f))
    
def main():
    bai1(x)
    bai2(2)
if __name__=="__main__":
    main()
```
