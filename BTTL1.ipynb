```python
import numpy as np
from numpy import cos,sin,pi

def hamcos(x):
    f=cos(x/pi)
    return f 

def hamsin(x):
  f=sin(x/3*pi)
  return f

def hambac2(a,b,c,x):
    f=a*x**2+b*x+c
    return f

def hambac3(a,b,c,d,x):
    f=a*x**3+b*x**2+c*x+d
    return f

def hambac1(a,b,x):
    f=a*x+b
    return f

x=np.linspace(-5,5,100000)

import matplotlib.pyplot as plt
%matplotlib inline

def bai1():
    y1=hambac1(1,1,x)
    y2=hambac2(2,0,0,x)
    y3=hambac3(2,2,-7,4,x)
    ig,ax=plt.subplots()
    ax.plot(x,y1)
    ax.plot(x,y2)
    ax.plot(x,y3)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.plot(x, y1,label=r'$y=x+1$')
    ax.plot(x, y2,label=r'$y=-2x^2 + 1x+5$')
    ax.plot(x, y3,label=r'$y=x^3+3x^2+5x+2$')
    ax.set_title("Bài 1")
    ax.legend()
    plt.show()
      
def bai2():
    y1=hamcos(x)
    y2=hamsin(x)
    fig,ax=plt.subplots()
    ax.plot(x,y1)
    ax.plot(x,y2)
    ax.set_xlabel(" x")
    ax.set_ylabel(" y")
    ax.plot(x, y1,label=r'$y=cos(x/pi)$')
    ax.plot(x, y2,label=r'$y=sin(pi*x/3)$')
    ax.set_title("Bài 2")
    ax.legend()
    plt.show()
    
def bai3():
    y1=hambac1(-5,3,x)
    y2=hambac2(2,-5,3,x)
    y3=hambac3(-2,-3,5,-9,x)
    y4=hamcos(x)
    y5=hamsin(x)
    fig, (ax) = plt.subplots(nrows=1, ncols=1)
    ax.plot(x,y1)
    ax.plot(x,y2)
    ax.plot(x,y3)
    ax.plot(x,y4)
    ax.plot(x,y5)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.plot(x,y1,label=r"y=−5x+3")
    ax.plot(x,y2,label=r"y=−2x^2−5x+3")
    ax.plot(x,y3,label=r"y=2x^3-3x^2+5x-9")
    ax.plot(x,y4,label=r"y=cos(x/pi)")
    ax.plot(x,y5,label=r"y=sin(pi*x/3)")
    ax.set_title("Bài 3")
    ax.legend()
    plt.show()



from matplotlib import cm

def ham3d(x,y):
    z=x*x+y*y -np.cos(12*x)-np.cos(18*y)
    return z

def bai4():
  x=np.linspace(-1,1,1000)
  y=np.linspace(-1,1,1000)
  x, y = np.meshgrid(x, y)
  z=ham3d(x,y)
  fig, ax =plt.subplots(subplot_kw={"projection":"3d"})
  rosen_surf = ax.plot_surface(x, y, z,cmap=cm.gist_heat_r, linewidth=45, antialiased=True)
  ax.set_zlim(-2,4)
  ax.set_xlabel("x")
  ax.set_ylabel("y")
  ax.set_zlabel("z")
  fig.colorbar(rosen_surf, shrink=0.7,aspect=50)
  ax.set_title("Bài 4")
  plt.show()
    
def main():
  bai1()
  bai2()
  bai3()
  bai4()   
if __name__=="__main__":
  main()
    
    
```
